```python
import pandas as pd
files =pd.read_excel('Financial Sample.xlsx')
print(files.head(5))
```

```python
files.columns
```

```python
files.index
```

```python
files.describe().head(3)
```

```python
# to find distinct countries
print(set(files['Country']))
```

```python
#data standardization 
files['Country'] = files["Country"].replace('United States of America','USA')
```

```python
# to select columns

files.groupby('Country')['Sale Price'].sum()
```

```python
# which country made the highest profit from sales
print(
    files.groupby('Country')['Profit']
    .sum()
    .round(2)
    .sort_values(ascending =False)
    
)
```

```python
# which country had  more units sold to
files.groupby('Country')['Units Sold'].sum().sort_values(ascending = False)
```

```python
print(set(files['Discount Band']))
```

```python
# to find which products and countries have a high discount band--total count of high discount band offered is 245
# which country had High discount band the most?
files[files['Discount Band'] == 'High'].groupby('Country').size().sort_values(ascending =False)
    
```

```python
# to find which products and countries have a low discount band--total count of high discount band offered is 245
# which country had the least discount band?
files[files['Discount Band'] == 'Low'].groupby('Country').size().sort_values(ascending =False)
    
```

```python
# which product made the highest profit and how many units sold
files.groupby("Product")[['Profit','Units Sold']].sum().round(2).sort_values(by= 'Profit', ascending = False)
```

```python
files[files['Product'] == 'Paseo'].groupby('Country').count().sort_values(ascending = False)
```
